```python
!pip install kagglehub

import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import os

path = kagglehub.dataset_download("kyanyoga/sample-sales-data")
print(path)

print(os.listdir(path))

csv_path = os.path.join(path, 'sales_data_sample.csv')
df = pd.read_csv(csv_path, encoding='latin1')

print(df.head())
print(df.shape)
print(df.columns.tolist())
print(df.isnull().sum())

grouped = df.groupby('PRODUCTLINE')['SALES'].sum()
print(grouped)

grouped.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

filtered = df[df['SALES'] > 5000]
print(filtered.head())

print(df.loc[0:2, ['PRODUCTLINE', 'SALES']])
print(df.iloc[0:2, 0:2])
```
